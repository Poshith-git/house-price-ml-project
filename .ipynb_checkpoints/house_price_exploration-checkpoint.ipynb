```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # For enhanced plots

# Load dataset
df = pd.read_csv('data/train.csv')  # Adjust path if needed

# Basic exploration
print("Dataset shape:", df.shape)  # Rows, columns
print("\nFirst 5 rows:\n", df.head())
print("\nColumn names (first 10):\n", df.columns[:10].tolist())
print("\nSummary stats for key columns:\n", 
      df[['SalePrice', 'LotArea', 'YearBuilt', 'OverallQual', 'GrLivArea']].describe())
```

```python
# Histogram of SalePrice
plt.figure(figsize=(8, 6))
sns.histplot(df['SalePrice'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of House Sale Prices')
plt.xlabel('Sale Price (USD)')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)
plt.savefig('sale_price_histogram.png', dpi=300, bbox_inches='tight')  # High-res save
plt.show()
```
